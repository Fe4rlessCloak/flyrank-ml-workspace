## 3. Success metric

*One metric you can defend. What number means 'good'?*

**Primary metric: Precision@K** — specifically Precision@20 and Precision@50.

**Precision@K measures:** "of the top K pages my model recommends for review, what fraction actually have a CTR problem?" This matches the real decision because an editor has limited capacity (e.g., 20 pages per week) and needs to know how many of those recommendations are worth their time.

**Ground truth definition:** A page is labeled as "truly underperforming on CTR" if its CTR falls below the 25th percentile of its position tier. For example, if pages at position tier "page_1" (positions 4-10) have a 25th percentile CTR of 0.12%, any page_1 page with CTR below 0.12% is considered underperforming. This definition is transparent, tunable, and grounded in the data.

**Secondary metric: RMSE** — measures how accurately the regression model predicts CTR on held-out data. 

**Baseline comparison:** I will compare the ML model's Precision@K against a simple hand-written rule: rank pages by CTR ascending, filtered to pages with at least 500 impressions. If the ML model achieves higher Precision@K, it proves that learning expected CTR from multiple features produces a better ranked queue than a single-threshold rule.

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

**Why ML beats fixed rules:** A simple rule like "flag pages with CTR < 0.5%" fails because it ignores position context. A page at position 2 with 0.5% CTR is severely underperforming, while a page at position 20 with 0.5% CTR is normal. A single threshold cannot capture this.

A more sophisticated rule could use position-tier thresholds, but it would still miss interactions between features.

ML handles these automatically by learning from data. A regression model can weigh all features simultaneously and find non-linear patterns that a hand-written rule would miss.
